```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Đọc dữ liệu
df = pd.read_csv("clean_data.csv")

# 2. Chỉ giữ các cột cần thiết
columns_to_keep = ['bath', 'balcony', 'total_sqft_float', 'bhk', 'price']
df_filtered = df[columns_to_keep].copy()

print(f"Shape trước khi xử lý NaN: {df_filtered.shape}")

# 3. Xóa các dòng có NaN (ở cột bath và balcony)
df_clean = df_filtered.dropna()

print(f"Shape sau khi xóa NaN: {df_clean.shape}")
print(f"Số dòng đã xóa: {df_filtered.shape[0] - df_clean.shape[0]}")
print(f"Tỷ lệ dữ liệu còn lại: {(df_clean.shape[0]/df_filtered.shape[0])*100:.2f}%")

# 4. Kiểm tra lại không còn NaN
print(f"\nKiểm tra NaN còn không: {df_clean.isnull().sum().sum() == 0}")
```

```python
X = df_clean.drop('price', axis=1)
y = df_clean['price']

print(f"\nShape của X: {X.shape}")
print(f"Shape của y: {y.shape}")

# 6. Chia train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=51
)

print(f"\nShape X_train: {X_train.shape}")
print(f"Shape X_test: {X_test.shape}")

# 7. Chuẩn hóa dữ liệu
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)

# 8. Huấn luyện mô hình
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

# 9. Đánh giá
y_pred = lr.predict(X_test_scaled)
r2_score = lr.score(X_test_scaled, y_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\n✅ KẾT QUẢ LINEAR REGRESSION:")
print(f"R² Score: {r2_score:}")
print(f"RMSE: {rmse:} Lakh")
```
